```python
import numpy as np

# Đọc và xử lý dữ liệu chiều cao từ file heights_1.txt
with open('heights_1.txt', 'r', encoding='utf-8-sig') as file:
    data = file.read()  # Đọc toàn bộ nội dung file
    # Loại bỏ dấu phẩy, tách các số, và chuyển thành float
    height_data = [float(num.replace(',', '')) for num in data.split()]

# Đọc và xử lý dữ liệu cân nặng từ file weights_1.txt
with open('weights_1.txt', 'r', encoding='utf-8-sig') as file:
    data = file.read()  # Đọc toàn bộ nội dung file
    # Loại bỏ dấu phẩy, tách các số, và chuyển thành float
    weight_data = [float(num.replace(',', '')) for num in data.split()]

# 1 và 2 Chuyển list height và weight thành numpy array
arr_height = np.array(height_data)
arr_weight = np.array(weight_data)

# Hiển thị kết quả
print("Chiều cao (inches):", arr_height)
print("Cân nặng (pounds):", arr_weight)

#3 
he_so_quy_doi_1=0.0254
arr_height_m=arr_height*he_so_quy_doi_1
print("he so sau ki quy doi la:", arr_height_m)

#4
he_so_quy_doi_2=0.453592
arr_weight_kg=arr_weight*he_so_quy_doi_2
print("he so sau khi quy doi la:",arr_weight_kg)

#5
can_nang=arr_weight_kg
chieu_cao=arr_height_m
bmi=can_nang/(chieu_cao*chieu_cao)
arr_bmi=bmi
print("bmi la",arr_bmi)

#6
index=50
gia_tri_can_nang=arr_weight_kg[index]
print("gia tri can nang o vi tri index 50 la:",gia_tri_can_nang)

#7
# Tạo mảng arr_height_m_100 gồm các phần tử từ index 100 đến 110 trong arr_height_m
arr_height_m_100 = arr_height_m[100:111]  # Chỉ số cuối là 111 để lấy cả phần tử ở index 110
print("arr_height_m_100:", arr_height_m_100)

#8
bmi_under_21=arr_bmi[bmi < 21]
print("cac cau thu bong chay co bmi duoi 21",bmi_under_21)

#9
# Tính chiều cao trung bình (tính theo mét)
average_height_m = np.mean(arr_height_m)
print("Chiều cao trung bình của các cầu thủ (m):", average_height_m)

# Tính cân nặng trung bình (tính theo kg)
average_weight_kg = np.mean(arr_weight_kg)
print("Cân nặng trung bình của các cầu thủ (kg):", average_weight_kg)

#10
# Tìm chiều cao lớn nhất (tính theo mét)
max_height_m = np.max(arr_height_m)
print("Chiều cao lớn nhất của các cầu thủ (m):", max_height_m)

# Tìm cân nặng lớn nhất (tính theo kg)
max_weight_kg = np.max(arr_weight_kg)
print("Cân nặng lớn nhất của các cầu thủ (kg):", max_weight_kg)

#11
# Tìm chiều cao lớn nhất (tính theo mét)
max_height_m = np.min(arr_height_m)
print("Chiều cao lớn nhất của các cầu thủ (m):", max_height_m)

# Tìm cân nặng lớn nhất (tính theo kg)
max_weight_kg = np.min(arr_weight_kg)
print("Cân nặng lớn nhất của các cầu thủ (kg):", max_weight_kg)




```
